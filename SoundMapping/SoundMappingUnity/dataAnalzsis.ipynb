```python
import pandas as pd
import numpy as np
```

```python
# Read the data from JSON file
data = pd.read_json('Assets/Data/datafirst.json')
```

```python
def getMag(x):
    return np.linalg.norm(np.array([x['x'], x['z']]))
```

```python
data['olfatiMag'] = data['olfatiData'].apply(lambda x: getMag(x))
data['obstacleMag'] = data['obstacleData'].apply(lambda x: getMag(x))
data['time'] = data.index * 1 / 30
```

```python
#plot all the magnitudes
data[0:1000].plot(x='time', y=['olfatiMag', 'obstacleMag'], title='Magnitude of olfati and obstacle data')
```

```python
data.plot(x='time', y=['olfatiMag', 'obstacleMag'], title='Magnitude of olfati and obstacle data')
```

```python
data['dOlfati'] = data['olfatiMag'].diff().fillna(0)
data['dObstacle'] = data['obstacleMag'].diff().fillna(0)
```

```python
data.plot(x='time', y=['dOlfati', 'dObstacle'], title='Difference of olfati and obstacle data')
```

```python
data[0:1000].plot(x='time', y=['dOlfati', 'dObstacle'], title='Difference of olfati and obstacle data')
```

```python
data[50:80].plot(x='time', y=['dOlfati', 'dObstacle'], title='Difference of olfati and obstacle data')
```

```python
#make mean window
data['dOlfatiMean'] = data['dOlfati'].rolling(window=10).mean()
data['dObstacleMean'] = data['dObstacle'].rolling(window=10).mean()
```

```python
data[50:80].plot(x='time', y=['dOlfatiMean', 'dObstacleMean'], title='smoothed Difference of olfati and obstacle data')
```

```python
data[0:1000].plot(x='time', y=['dOlfatiMean', 'dObstacleMean'], title='smoothed Difference of olfati and obstacle data')
```
